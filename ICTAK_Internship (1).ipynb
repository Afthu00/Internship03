```python
def outlierremover(data):
    Q1=data.quantile(0.25)
    Q3=data.quantile(0.75)
    IQR=Q3-Q1
    LowerBound=Q1-1.5*IQR
    UpperBound=Q3+1.5*IQR
    return data[(data >= LowerBound) & (data <= UpperBound)]
```

```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col] = outlierremover(df[col])
```

```python
def print_evaluation(model_name, y_true, y_pred):
    print(f"--- {model_name} Evaluation ---")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print("Classification Report:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    print("\n")
```

```python
print_evaluation("KNN", y_test, y_pred_knn)
print_evaluation("Random Forest", y_test, y_pred_rf)
print_evaluation("Logistic Regression", y_test, y_pred_lr)
```

```python
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
```

```python
y_prob_knn = knn_best.predict_proba(X_test)
y_prob_rf = rf_best.predict_proba(X_test)
y_prob_lr = lr_best.predict_proba(X_test)
```

```python
def compute_roc_auc(y_test_bin, y_prob, model_name):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(y_test_bin.shape[1]):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    return fpr, tpr, roc_auc
```

```python
fpr_knn, tpr_knn, auc_knn = compute_roc_auc(y_test_bin, y_prob_knn, 'KNN')
fpr_rf, tpr_rf, auc_rf = compute_roc_auc(y_test_bin, y_prob_rf, 'Random Forest')
fpr_lr, tpr_lr, auc_lr = compute_roc_auc(y_test_bin, y_prob_lr, 'Logistic Regression')
```

```python
plt.figure(figsize=(10, 8))
plt.plot(fpr_knn["micro"], tpr_knn["micro"], label=f'KNN (AUC = {auc_knn["micro"]:.4f})')
plt.plot(fpr_rf["micro"], tpr_rf["micro"], label=f'Random Forest (AUC = {auc_rf["micro"]:.4f})')
plt.plot(fpr_lr["micro"], tpr_lr["micro"], label=f'Logistic Regression (AUC = {auc_lr["micro"]:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
models = ['KNN', 'Random Forest', 'Logistic Regression']
accuracies = [0.8145, 0.8455, 0.5684]
f1_scores = [0.81, 0.85, 0.47]
x = np.arange(len(models))
width = 0.35
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, accuracies, width, label='Accuracy', color='b')
bars2 = ax.bar(x + width/2, f1_scores, width, label='F1 Score', color='g')

ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
